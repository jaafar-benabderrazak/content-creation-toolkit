```python
# Clean and format the DataFrame
# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Set display options for better formatting
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

# Display formatted DataFrame
print("DataFrame Info:")
print(f"Shape: {df.shape}")
print(f"Columns: {len(df.columns)}")
print("\nFirst 3 rows:")
print(df.head(3).to_string())

# Or display specific columns only
key_cols = ['Intitulé Client', 'Nom', 'Prenom', 'Fonction', 'Ville']
print(f"\nKey columns preview:")
print(df[key_cols].head().to_string())

# Save to Excel with metadata
with pd.ExcelWriter('cleaned_data.xlsx', engine='openpyxl') as writer:
    # Create metadata sheet
    metadata = pd.DataFrame({
        'Property': ['File Name', 'Total Rows', 'Total Columns', 'Date Processed', 'Original Header Row'],
        'Value': ['cleaned_data.xlsx', df.shape[0], df.shape[1], pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S'), '8th row (index 7)']
    })
    metadata.to_excel(writer, sheet_name='Metadata', index=False)
    
    # Save main data
    df.to_excel(writer, sheet_name='Data', index=False)
```

```python
import pandas as pd

# Read Excel file with 8th row as header (skip first 7 rows)
df = pd.read_excel('contactes bensaid Data.xlsx', header = 7)
# Clean and format the DataFrame
# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Set display options for better formatting
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

# Display formatted DataFrame
print("DataFrame Info:")
print(f"Shape: {df.shape}")
print(f"Columns: {len(df.columns)}")
print("\nFirst 3 rows:")
print(df.head(3).to_string())

# Or display specific columns only
key_cols = ['Intitulé Client', 'Nom', 'Prenom', 'Fonction', 'Ville']
print(f"\nKey columns preview:")
print(df[key_cols].head().to_string())

# Save to Excel
df.to_excel('cleaned_data.xlsx', index=False)

# Group by Secteur and save each group to separate Excel sheets
with pd.ExcelWriter('clients_by_secteur.xlsx', engine='openpyxl') as writer:
    for secteur, group in df.groupby('Secteur'):
        # Sort by Intitulé Client
        group_sorted = group.sort_values('Intitulé Client')
        # Save to sheet (clean sheet name)
        sheet_name = str(secteur)[:31]  # Excel sheet names max 31 chars
        group_sorted.to_excel(writer, sheet_name=sheet_name, index=False)
```
